# EV Market Analysis

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# 1st data Importation with functions

```python
df1 = pd.read_excel("3.EV-Market-India-Dataset.xlsx")
```

```python
# Number of rows & columns in the dataset
df1.shape
```

```python
# Top 5 records of the dataset
df1.head()
```

```python
# Bottom 5 records of the dataset 
df1.tail()
```

```python
# Overall information about the dataset
df1.info()
```

```python
# Checking if there is any null values is present in the dataset or not
pd.isnull(df1)
```

```python
# Number of null values in the columns
pd.isnull(df1).sum()
```

```python
# Column headers
df1.columns
```

```python
# Describe() method returns description of the data in the DataFrame (i.e count,mean,std,min..etc)
df1.describe()
```

# Exploratory Data Analysis

```python
# Column Headers
df1.columns
```

```python
# Bar Plot
s1 = sns.countplot(x='Segment',data=df1)
```

```python
# Bar Plot with their count
s1 = sns.countplot(x='Segment',data=df1)
for bars in s1.containers:
    s1.bar_label(bars)
```

# Group By function

```python
# Groupby() function in pandas is used to group data based on one or more columns in the DataFrame
```

```python
df1.groupby(['Segment'], as_index=False)['PriceEuro'].sum().sort_values(by='PriceEuro', ascending=False)
```

```python
# Bar Plot with groupby() fuction
g1 = df1.groupby(['Segment'], as_index=False)['PriceEuro'].sum().sort_values(by='PriceEuro', ascending=False)
sns.barplot(x='Segment', y='PriceEuro', data=g1)
```

```python
# Histogram
df1['Segment'].hist()
```

```python
# Ploting histogram by changing the datatype of the column
df1['Brand'] = df1 ['Brand'].astype(str)
column_data = df1['Brand']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=100, edgecolor='Yellow')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter Plot
x_data = df1['Seats']
y_data = df1['BodyStyle']
plt.scatter(x_data, y_data)
plt.xlabel('Seats')
plt.ylabel('BodyStyle')
plt.title('Info')
plt.show()
```

```python
# Count of Models by Brands
plt.figure(figsize=(12,6))
sns.countplot(data=df1, x='Brand')
plt.xlabel('Brand')
plt.ylabel('Model')
plt.title('Top Brands in the EV Market with their model count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 10 Brands
top_10_brand = df1['Brand'].value_counts().head(10)
# Count of Models by Brands
plt.figure(figsize=(12,6))
sns.countplot(data=df1[df1['Brand'].isin(top_10_brand.index)], x='Brand')
plt.xlabel('Brand')
plt.ylabel('Model')
plt.title('Top Brands in the EV Market with their model count')
plt.xticks(rotation=45)
plt.show()
```
